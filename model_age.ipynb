SVM

```python
from sklearn.model_selection import train_test_split
from sklearnex import patch_sklearn
from sklearnex.svm import SVC
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(utkf_images, utkf_ages, test_size=0.3, random_state=42)

# Enable Intel optimizations for scikit-learn
patch_sklearn()

# Create an instance of IntSVC
clf = SVC(kernel='rbf', C=1.0, gamma='scale')

# Fit the model on the training data
clf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = clf.predict(X_test)

# Evaluate the model performance
score = clf.score(X_test, y_test)
print(f"Accuracy: {score}")
```

```python
import cv2
import matplotlib.pyplot as plt

import pickle

with open('model_svc2.pkl', 'rb') as f:
    clf = pickle.load(f)

# imagePath = "C:/Users/ayoub/Downloads/"
imagePath = "Images/3.jpg"
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
img = cv2.imread(imagePath)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
    face = img[y:y+h,x:x+w]
    face = cv2.resize(face,(65,65))
    face = cv2.cvtColor(face,cv2.COLOR_RGB2GRAY)
    face = face.flatten()
    result = clf.predict([face])
    if result == 1 : result = 'Kid'
    if result == 2 : result = 'Ado'
    if result == 3 : result = 'Jeun'
    if result == 4 : result = 'Adult'
    # if result == 5 : result = 'Adult 2'
    if result == 5 : result = 'Old'
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    font = cv2.FONT_HERSHEY_SIMPLEX
    fontScale = 1
    org = (x, y-30)
    color = (0, 255, 0)
    thickness = 1
    cv2.putText(img, str(result), org, font, fontScale, color, thickness)
cv2.imshow('Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
